```python
import fastai
from PIL import Image
import matplotlib.pyplot as plt
import pathlib
```

```python
temp = pathlib.PosixPath
pathlib.PosixPath = pathlib.WindowsPath

# Import fastai & load the learner
from fastai.vision.all import *
learn_inf = load_learner("Scuba_Gesture_Detection_NN.pkl")
```

```python
# Restore the original PosixPath
pathlib.PosixPath = temp

# Provide a real file path "Insert imagr path", which you want to classify

#Below code to verify file from dataset
#image_path = r"CADDY_gestures_complete_v2_release\brodarski-A\true_positives\raw\brodarski-A_00018_left.jpg"

# Tested result for a random image not from dataset
image_path = r"Unseen_gesture_not_from_dataset.jpg"
img = PILImage.create(image_path)


# Predict
pred_class, pred_idx, probs = learn_inf.predict(img)
print("Predicted class:", pred_class)
print("Probability:", probs[pred_idx])
plt.imshow(img)
plt.axis('off')
plt.show()
```
